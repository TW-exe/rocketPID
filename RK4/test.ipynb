```python
#libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rk4singlestep(func,dt,tk,xk):

    
    f1 = func(tk,xk)
    f2 = func(tk+(dt/2),xk +(dt/2)*f1)
    f3 = func(tk+(dt/2),xk +(dt/2)*f2)
    f4 = func(tk+(dt),xk +(dt)*f3)

    xout = xk + (dt/6)*(f1 + 2*f2 + 2*f3 + f4)

    return xout
```

```python
def lorenz(t,x):
     # x is a 3D state vector
      sigma = 10
      beta = 8/3
      rho = 28

      dx = [sigma*(x[1]-x[0]), 
          x[0]*(rho - x[2])-x[1], 
          x[0]*x[1]-beta*x[2]]
    
      return np.array(dx)
```

```python
#lorenz parameters 
sigma = 10
beta = 8/3
rho = 28



#intial condition
x0 = [-8,8,27]

#tragectory 
dt = 0.01
T=10
num_time_pts = int(T/dt)
#t is the time step array
t = np.linspace(0,T,num_time_pts)

X = np.zeros((3,num_time_pts))

X[:, 0] = x0
```

```python
xin = x0

for i in range(num_time_pts-1):
    xout = rk4singlestep(lorenz,dt,t[i],xin)
    X[:,i+1] = xout
    xin = xout
```

```python
ax = plt.figure().add_subplot(projection ='3d')
ax.plot(X[0,:],X[1,:],X[2,:],'b')
plt.show
```
